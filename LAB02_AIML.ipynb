```python
problem 1 : Bar plot
Task : Create a bar plot showing the number of students in different classes.

Data:
-Classes:['Class A','Class B','Class C','Class D']
-Number of students : [23,17,30,25]

Instructions:
1.Label the x-axis as "Classes".

2. Label the y-axis as "Number of Students".

3. Add a title: "Number of Students in Each Class".

4. Choose different colors for each bar.
```

```python
import matplotlib.pyplot as plt

classes = ['Class A', 'Class B', 'Class C', 'Class D']
number_of_students = [23, 17, 30, 25]

plt.bar(classes, number_of_students, color=['pink', 'blue', 'red', 'yellow'])
plt.xlabel('Classes')
plt.ylabel('Number of Students')
plt.title('Number of Students in Each Class')

plt.show()
```
